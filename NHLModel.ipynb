```python
import sklearn
```

```python
import pandas as pd
df = pd.read_csv('NHLSchedule.csv')
df.head()
```

```python
df1 = pd.read_csv('HomeTeamStats.csv')
df1.head()
```

```python
df2 = pd.merge(df, df1, left_on=['Home'], right_on=['HTm'], how="left")
df2.head()
```

```python
df3 = pd.read_csv('VisitorTeamStats.csv')
df3.head()
```

```python
df4 =  pd.merge(df2, df3, left_on=['Visitor'], right_on=['VTm'], how="left")
df4.head()
```

```python
df4.to_csv('TrainData.csv')
```

```python
df5 = pd.read_csv('TrainData.csv')
```

```python
df_output = df5[['Result','Gtotal']]
df_output.head()
```

```python
df_stats =  df5[['HRk','HAvAge','HGF/G','HGA/G','HPP','HPPO','HPP%','HPPA','HPPOA','HPK%','HSH','HSHA',
             'HPIM/G','HoPIM/G','HS','HS%','HSA','HSV%','HSO',
             'VRk','VAvAge','VGF/G','VGA/G','VPP','VPPO','VPP%','VPPA','VPPOA','VPK%','VSH','VSHA',
             'VPIM/G','VoPIM/G','VS','VS%','VSA','VSV%','VSO']]
df_stats.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
predictor = LinearRegression(n_jobs=-1)

predictor.fit(X=df_stats, y=df_output)
```

```python
df6 = pd.read_csv('TestData.csv')
df6.head()
```

```python
df7 = pd.merge(df6, df1, left_on=['Home'], right_on=['HTm'], how="left")
df7.head()
```

```python
df8 = pd.merge(df7, df3, left_on=['Visitor'], right_on=['VTm'], how="left")
df8.head()
```

```python
df8.to_csv('TestDataFull.csv')
```

```python
df_test = df8[['HRk','HAvAge','HGF/G','HGA/G','HPP','HPPO','HPP%','HPPA','HPPOA','HPK%','HSH','HSHA',
             'HPIM/G','HoPIM/G','HS','HS%','HSA','HSV%','HSO',
             'VRk','VAvAge','VGF/G','VGA/G','VPP','VPPO','VPP%','VPPA','VPPOA','VPK%','VSH','VSHA',
             'VPIM/G','VoPIM/G','VS','VS%','VSA','VSV%','VSO']]
df_test.head()
```
